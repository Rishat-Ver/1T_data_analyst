# Импорты

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

# А/А тестирование

## Шаг 1: генерация данных

```python
# # Устанавливаем начальное значение для генератора случайных чисел для обеспечения воспроизводимости
np.random.seed(42)

# Генерируем массив из 1000 случайных чисел для группы A, которые следуют нормальному распределению
# с средним (loc) 50 и стандартным отклонением (scale) 10
group_a = np.random.normal(loc=50, scale=10, size=1000)

# Генерируем массив из 1000 случайных чисел для группы B
group_b = np.random.normal(loc=50, scale=10, size=1000)
```

## Шаг 2: визуализация данных

```python
# Устанавливаем размер фигуры
plt.figure(figsize=(8, 6))

# Строим гистограмму для группы A
plt.hist(group_a, bins=30, alpha=0.5, label='Группа A')

# Строим гистограмму для группы B
plt.hist(group_b, bins=30, alpha=0.5, label='Группа B')

# Устанавливаем метку по оси x
plt.xlabel('Значения')

# Устанавливаем метку по оси y
plt.ylabel('Частота')

# Устанавливаем заголовок графика
plt.title('Гистограмма двух контрольных групп')

# Добавляем легенду для различения групп
plt.legend()

# Отображаем график
plt.show()
```

## Шаг 3: проведение t-теста

```python
# Устанавливаем уровень значимости для теста
alpha = 0.05

# Выполняем t-критерий Стьюдента для независимых выборок между группой A и группой B
t_stat, p_value = ttest_ind(group_a, group_b)
```

## Шаг 4: вывод результатов анализа

```python
# Выводим значение p-value с точностью до 4 знаков после запятой
print(f'Значение p-value: {p_value: .4f}')

# Проверяем, больше ли значение p-value порогового уровня значимости alpha
if p_value > alpha:
    print('Статистичиески значимых различий между группами не обнаружено.' )
else:
    print('Обнаружены статистические значимые различия между группами.' )
```
