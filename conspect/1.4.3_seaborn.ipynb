```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

## <center>Pairplot<center>

```python
sns.set_theme(style="whitegrid")

df = sns.load_dataset("penguins")
sns.pairplot(df, hue="island")
```

```python
# загрузка набора данных tips
tips = sns.load_dataset("tips")

# построение графика lmplot, показывающего связь между суммой чаевых и общей суммой счета
sns.lmplot(x="total_bill", y="tip", data=tips, height=10, aspect=2)

# установка заголовка графика
plt.title("Tips by Total Bill")

# отображение графика
plt.show()
```

## <center>Точечная диаграмма - Scatterplot<center>

```python
# загружаем набор данных tips
tips = sns.load_dataset("diamonds")

# создаем scatterplot
f, ax = plt.subplots(figsize=(10, 10))
sns.scatterplot(x="carat", y="price", data=tips)

# добавляем заголовок
plt.title("Diamonds")

# отображаем график
plt.show()
```

```python
sns.set_theme(style="whitegrid")

# Загрузим датасет из Seaborn
diamonds = sns.load_dataset("diamonds")

# Нарисуем точечную диаграмму

f, ax = plt.subplots(figsize=(10, 10))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=diamonds, ax=ax);
#plt.title('Scatterplot', fontsize = 20)
#plt.xlabel('carat', fontsize = 16)
#plt.ylabel('price', fontsize = 16)
```

```python
diamonds.head()
```

<div class="alert alert-info"><b>Примечание</b>
    
<ul><li> 
    set_theme() - позволяет установить аспекты визуальной темы для всех графиков, она применена по умолчанию с определенными параметрами, однако их можно редактировать </li>
<li>despine() - Позволяет обрезать график по сторонам, то есть срезать "шипы"</li></ul></div>


<div class="alert alert-info"><b>Примечание</b>
    
<ul><li> 
    hue -  может принимать категориальный столбец данных для разделения элементов на графике. Примерами могут служить признаки, такие как пол, регион, группа, категория, временной период и т.д. </li>
<li>size - данный параметр определяет размер точек на диаграмме в соответствии с выбранным столбцом</li></ul>
<li>palette - Определяет цветовую палитру у графика</li></ul></div>

<div class="alert alert-info">
    
**Диаграмма рассеяния** - это способ создания графика, который показывает распределение точек в двумерном пространстве на основе двух числовых переменных набора данных. Каждая точка представляет собой значение одной переменной на оси x и значение другой переменной на оси y. Данная диаграмма позволяет наглядно представить связь или ее отсутствие между двумя переменными.,<br><br> В функции `relplot()` этот тип графика является одним из возможных типов (изменяется с помощью параметра `kind`), что упрощает создание таких графиков в сочетании с другими типами графиков, такими как `lineplot()`и `scatterplot()`.<br><br> Кроме того, вместо параметра `figsize` в `relplot()` можно использовать параметры `height` и `aspect` для настройки размера и формы графика.

## <center>"Ящик с усами" - Boxplot<center>

```python
sns.set_theme(style="ticks", palette="pastel")


tips = sns.load_dataset("tips")

f, ax = plt.subplots(figsize=(10, 10))
sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["b", "g"],
            data=tips)
sns.despine(offset=10, trim=True)
```

```python
tips.head()
```

```python
sns.set_theme(style="ticks")


f, ax = plt.subplots(figsize=(10, 10))
ax.set_xscale("log")
sns.despine(f, left=True, bottom=True)



planets = sns.load_dataset("planets")


sns.boxplot(x="distance", y="method", data=planets,
            whis=[0, 100], width=.6, palette="vlag")


sns.stripplot(x="distance", y="method", data=planets,
              size=4, color=".3", linewidth=0)

# Tweak the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```

```python
planets.head()
```

```python
tips = sns.load_dataset("tips")

# построение графика catplot с группировкой по полу и времени обеда
g = sns.catplot(x="sex", y="total_bill", hue="time", data=tips, kind="box", height=6, aspect=1.2)

# добавление заголовка графика
g.fig.suptitle("Total Bill by Gender and Time of Day")

# отображение графика
plt.show()
```

<div class="alert alert-info">

Функция `catplot()` в Seaborn позволяет объединить в себе возможности нескольких функций, таких как `boxplot()`, `violinplot()`, `boxenplot()`, `stripplot()`, `swarmplot()`, `pointplot()`, `barplot()` и `countplot()`.<br><br> Переключение между этими типами графиков происходит с помощью параметра `kind`, что упрощает создание графиков и требует меньше запоминания разных названий функций и параметров.<br><br> Кроме того, функция `catplot()` позволяет настроить размер и форму графика с помощью параметров `height` и `aspect`, и поэтому `figsize` больше не нужен.

## <center>Гистограммы - Hist<center>

```python
sns.set_theme(style="darkgrid")
df = sns.load_dataset("penguins")
sns.displot(
    df, x="flipper_length_mm", col="species", row="sex",
    binwidth=3, height=3, facet_kws=dict(margin_titles=True),kind='hist'
)
```

```python
sns.set_theme(style="ticks")

diamonds = sns.load_dataset("diamonds")

f, ax = plt.subplots(figsize=(12, 10))
sns.despine(f)

sns.histplot(
    diamonds,
    x="price", hue="cut",
    multiple="stack", # 'dodge' для примера
    palette="light:m_r",
    edgecolor="0.1",
    linewidth=.5,
    log_scale=True,
)
ax.xaxis.set_major_formatter(mpl.ticker.ScalarFormatter())
ax.set_xticks([500, 1000, 2000, 5000, 10000])
```

```python
diamonds.head()
```

<div class="alert alert-info">
    
Функция `displot()` в Seaborn объединяет возможности функций `histplot()`, `kdeplot()` и `ecdfplot()`, позволяя использовать различные типы графиков для визуализации распределения одной или нескольких переменных, которые можно переключать с помощью параметра `kind`.<br><br> Это уменьшает необходимость запоминать различные названия функций и параметров и упрощает создание графиков. Кроме того, `displot()` позволяет настроить размер и форму графика с помощью параметров `height` и `aspect`, и поэтому `figsize` больше не нужен.

## <center>Тепловая карта - Heatmap<center>

```python
sns.set_theme()


flights_long = sns.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")


f, ax = plt.subplots(figsize=(12, 12))
sns.heatmap(flights, annot=True, fmt="d", linewidths=0.3, ax=ax, cmap='BuPu');
```

## <center>Столбчатые диаграммы - Barplot<center>

```python
sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(6, 15))


crashes = sns.load_dataset("car_crashes").sort_values("total", ascending=False)

sns.set_color_codes("pastel")
sns.barplot(x="total", y="abbrev", data=crashes,
            label="Total", color="b")


sns.set_color_codes("muted")
sns.barplot(x="alcohol", y="abbrev", data=crashes,
            label="Alcohol-involved", color="b")


ax.legend(ncol=1, loc="upper right", frameon=True)
ax.set(xlim=(0, 24), ylabel="",
       xlabel="Automobile collisions per billion miles")
sns.despine(left=True, bottom=True)
```

```python
f, ax = plt.subplots(figsize=(15, 6))


crashes = sns.load_dataset("car_crashes").sort_values("total", ascending=False)


sns.set_color_codes("pastel")
sns.barplot(x="abbrev", y="total", data=crashes,
            label="Total", color="b")


sns.set_color_codes("muted")
sns.barplot(x="abbrev", y="alcohol", data=crashes,
            label="Alcohol-involved", color="b")


ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 24), xlabel="Automobile collisions per billion miles",
       ylabel="")
sns.despine(left=True, bottom=True)
```
