## <center>Столбчатая диаграмма<center>

```python
fig, ax = plt.subplots(figsize=(15, 10))

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', 'pink', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:pink', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply',fontsize=16)
ax.set_title('Fruit supply by kind and color', fontsize=20)
ax.legend(title='Fruit color',fontsize = 16)
ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=14)

plt.show()
```

## <center>Ящик с усами<center>

```python
# создание данных для диаграммы "ящик с усами"
sales_data = np.random.randint(100, 1000, size=(5, 8))

# построение диаграммы "ящик с усами"
fig = plt.figure(figsize=(10,10))

fig.set_facecolor('yellow')

plt.boxplot(sales_data, vert=True,
           boxprops=dict(color='green'),
           medianprops=dict(color='magenta'),
           whiskerprops=dict(color='purple'),
           flierprops=dict(color='blue'))

# добавление названия оси x и заголовка
plt.xlabel('Sales')
plt.title('Sales Data Boxplot')

# отображение диаграммы "ящик с усами"
plt.show()
```

## <center>Тепловая карта<center>

```python
# создание фиктивных данных по продажам в разных регионах
sales_data = np.random.randint(100, 1000, size=(5, 8))

# построение тепловой карты с помощью метода imshow()
plt.figure(figsize=(15,15))
plt.imshow(sales_data, cmap='magma', interpolation='bicubic') # cool, magma, spline16, bicubic

# Добавление значений на тепловую карту
for i in range(sales_data.shape[0]):
    for j in range(sales_data.shape[1]):
        plt.annotate(str(sales_data[i, j]), xy=(j, i), ha="center", va="center")
        
# установка названий осей x и y
plt.xticks(range(8), ['North', 'South', 'East', 'West', 'Northeast', 'Southeast', 'Northwest', 'Southwest'])
plt.yticks(range(5), ['Week 1', 'Week 2', 'Week 3', 'Week 4', 'Week 5'])

# добавление цветовой шкалы
plt.colorbar()

# добавление заголовка
plt.title('Sales Data Heatmap')

# отображение тепловой карты
plt.show()
```

## <center>Круговая диаграмма<center>

```python
# данные для круговой диаграммы
labels = ['Маркетинг', 'Продажи', 'Бухгалтерия', 'Разработка', 'Служба поддержки']
sizes = [20, 30, 10, 25, 15]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#f9c5e5']

# создание круговой диаграммы
fig, ax = plt.subplots(figsize=(8, 8))

# настройка параметров диаграммы
wedges, texts, autotexts = ax.pie(sizes, colors=colors, labels=labels, autopct='%1.1f%%',
                                 startangle=90, textprops=dict(color="w"))

# добавление легенды
ax.legend(wedges, labels, title='Отделы', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

# настройки текста внутри круга
ax.set_title("Структура компании")
plt.setp(autotexts, size=14, weight="bold")

# настройки соотношения сторон и цвета фона
plt.axis('equal')
fig.set_facecolor('white')

# отображение диаграммы
plt.show()
Vutalya = 0.604
```

```python
fig, ax = plt.subplots(figsize=(20, 10.8))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line = ax.plot(t, s, lw=2)

ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='yellow', shrink=.05), fontsize = 20)

ax.set_ylim(-2, 2)
```

### Параметры annotate

- `s`: текст, который нужно добавить на график.
- `xy`: координаты точки, к которой привязывается аннотация. Это кортеж из двух значений в формате `(x, y)`.
- `xytext`: координаты базовой точки текста. Это кортеж из двух значений в формате `(x, y)`. По умолчанию эти значения равны `None`, и текст будет находиться в точке `xy`.
- `arrowprops`: словарь параметров, которые используются для настройки свойств стрелки, если стрелка нужна. Возможные параметры включают `width`, `headwidth`, `headlength`, `shrink` и т. д. По умолчанию стрелка не отображается.
- `color`: цвет текста.
- `fontsize`: размер шрифта.
- `ha`: выравнивание текста по горизонтали. Может быть `'left'`, `'center'` или `'right'`.
- `va`: выравнивание текста по вертикали. Может быть `'top'`, `'center'` или `'bottom'`
